```python
df = df.drop_duplicates()
```

```python

df["CCAvg"] = pd.to_numeric(df["CCAvg"], errors="coerce")


df["CCAvg"].fillna(df["CCAvg"].mean(), inplace=True)


df = df.drop(columns=["ID", "ZIP Code"], errors="ignore")



```

```python

df = pd.get_dummies(df, columns=["Education"], prefix="Edu", drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numeric_cols = ["Age", "Experience", "Income", "CCAvg", "Mortgage"]

df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
```

```python
print(df.info())
print(df.head())
```

```python
df.to_csv("bank_loan_clean_ml.csv", index=False)
print("Data cleaned. New shape:", df.shape)
```
